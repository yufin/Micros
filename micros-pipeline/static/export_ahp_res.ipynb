```python
import pandas as pd
import  requests
import json
from pprint import pprint
```

```python
df = pd.read_csv("Result_16.csv")
```

```python
df
```

```python
def get_ahp(claim_id: int, token: str):
    url = "http://10.0.232.22:8009/micros/rc/v3/report/ahp?claim_id={}".format(claim_id)
    headers = {'Authorization': f'Bearer {token}'}
    r = requests.get(url, headers=headers)
    return r.json()
    
for i, s in df.iterrows():
    print(i)
    cid = s['claim_id']
    res = get_ahp(cid, "78cc850dc4bc4c8bb01ec54dce10a566")
    # print(res)
    try:
        score = res['data']['l4']['score']
        res_json = json.dumps(res["data"])
        df.loc[i, '总分'] = float(score) * 10
        # df.loc[i, 'result_json'] = res_json
    except Exception as e:
        raise e
    
df.to_csv("Result_ahpScore.csv")
```

```python
def gen_df(res_detail: dict, file_name: str):
    s0 = pd.Series(res_detail['l0'])
    s1 = pd.Series(res_detail['l1'])
    df_detail = pd.concat([s0, s1], axis=1)
    df_detail.to_csv("./details/"+file_name)

for i in range(len(df)):
    print(i)
    s = df.iloc[i]['result_json']
    d = json.loads(s)
    fname = str(i) + ".csv"
    gen_df(d, fname)
```
